```python
import numpy as np
import math
```

```python
chars = np.array(['diluc', 'kooching', 'keqing', 'ayato', 'nilou', 'amber', 'childe_phys'])
```

```python
permuted = {}
```

```python
for i in range (math.factorial(len(chars))):
    rand_indices = np.random.choice(a=7, replace=False, size=7)
    permuted[i] = rand_indices
```

```python
finalPerm = {}
```

```python
for i in permuted:
    finalPerm[i] = [chars[permuted[i][j]] for j in range (0, len(chars))]
```
