```python
X = 32
means = [26, 28, 30, 32, 34, 36, 38]
# Sds = [1, 2, 3, 4, 5, 6]
sd = 2
L = []

for i in range (0, len(means)):
    l = np.log(norm.pdf(X, means[i], sd))
    L.append(l)

print (L)
```

```python
from scipy.optimize import minimize
```

```python
np.random.seed(123)
mean1 = 2
sd1 = 3
N = 100
X = np.random.normal(mean1, sd1, (N,)) #this is what is given
p = [1, 1]
def maxL(p, X):
    mu = p[0]
    sd = p[1]
    l = np.sum(np.log(norm.pdf(X, mu, sd)))
    return -l
print (maxL(p, X))

def cons(p):
    return p

con = {'type': 'ineq', 'fun':cons}
minimize(maxL, p, args=(X, ), constraints = con)
```
