+ VIF = Variability Inflation Factor
+ Used to check correlation between 2 variables
+ Suppose there are 3 variables x1, x2, x3, y; y dependent
+ formula : 1/(1-R^2)
+ Mechanism: Python iteratively figures out VIF using one xn as dependent, others as independent
