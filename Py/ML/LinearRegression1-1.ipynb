```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("tvmarketing.csv")
print (df)
```

```python
X = df["TV"]
Y = df["Sales"]
w = np.zeros(len(X))
b = np.zeros(len(X))
LR = 0.00001
it = 10000
J = []
```

```python
for x in range(it):
    Yhat = (w*X) + b
    DiffWRTw = (2/float(len(X)))*((Yhat-Y)*X)
    DiffWRTb = (2/float(len(X)))*((Yhat-Y))
    w = w-sum((LR*DiffWRTw))
    b = b-sum((LR*DiffWRTb))
```

```python
plt.scatter(X, Y)
plt.plot([min(X), max(X)], [min(Yhat), max(Yhat)], color="red")
plt.show()
```
