```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Windmill1.csv")
X = df["Wind"]
Y = df["Power"]
print (Y.isna().sum())
print (X.isna().sum())
```

```python
w = 0
b = 0
LR = 0.0001
it = 100000
J = np.zeros(it)
```

```python
for x in range (0, it):
    Yhat = (w*X)+b
    J1 = (1/(2*len(X)))*(((Yhat-Y)**2).sum())
    J[x] = J1
    DiffWRTw = (2/float(len(X)))*((X*(Yhat-Y)).sum())
    DiffWRTb = (2/float(len(X)))*((Yhat-Y).sum())
    w = w-(LR*DiffWRTw)
    b = b-(LR*DiffWRTb)

print (w, b)
```

```python
it1 = list(range(it))
print (len(it1), len(J))
plt.plot(it1, J)
plt.show()
```
