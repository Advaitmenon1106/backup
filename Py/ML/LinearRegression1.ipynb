```python
X = df["TV"]
Y = df["Sales"]
X = X/max(X)
Y = Y/max(Y)
```

```python
w = 0
b = 0
LR = 0.003
it = 10000
J = []
for x in range(it):
    Yhat = (w*X) + b
    J1 = (1/(2*len(X)))*(((Yhat-Y)**2).sum())
    J.append(J1)
    DiffWRTw = (1/float(len(X)))*((X*(Yhat-Y)).sum())
    DiffWRTb = (1/float(len(X)))*((Yhat-Y).sum())
    w = w-(LR*DiffWRTw)
    b = b-(LR*DiffWRTb)
print (w, b)
```

```python
plt.scatter(X, Y)
plt.plot([min(X), max(X)], [min(Yhat), max(Yhat)], color="red")
plt.show()
```
