```python
import numpy as np
import pandas as pd
import seaborn as sns
import statsmodels.api as smf
import statsmodels.formula.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor
from statsmodels.tools.tools import add_constant
```

```python
auto = pd.read_csv('Price_Prediction.csv', na_values='?')
auto.head()
```

```python
auto.isna().sum()
```

```python
auto1 = auto.dropna()
```

```python
auto1.reset_index(inplace=True)
```

```python
auto1.drop(columns='index', inplace=True)
```

```python
autoQuant = auto1.drop(columns=['symboling', 'make', 'fuel_type', 'aspiration', 'num_of_doors', 'body_style', 'drive_wheels', 'engine_location', 'engine_type', 'num_of_cylinders', 'fuel_system', 'price'])
```

```python
autoQuant
```

```python
X1 = add_constant(autoQuant)
vif = pd.DataFrame()
vif['Features'] = X1.columns
vif['VIF'] = [variance_inflation_factor(X1.values, i) for i in range (len(X1.columns))]
```
