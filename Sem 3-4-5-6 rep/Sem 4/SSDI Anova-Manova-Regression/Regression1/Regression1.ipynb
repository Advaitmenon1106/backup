```python
import numpy as np
import pandas as pd
import statsmodels.api as smf
import statsmodels.formula.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
corn = pd.read_csv('Corn.csv')
corn
```

```python
sns.scatterplot(data=corn, x=corn.nitrogen, y=corn.weight)
```

```python
fit = sm.ols('weight~nitrogen', data= corn).fit()
fit.summary()
```
