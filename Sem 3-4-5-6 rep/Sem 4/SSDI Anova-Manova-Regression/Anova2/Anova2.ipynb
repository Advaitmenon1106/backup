```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols
from statsmodels.stats.multicomp import pairwise_tukeyhsd
```

```python
df = pd.read_csv("Anova Dataset 2.csv")
df
```

```python
df2 = pd.melt(df, value_vars=["Education", "Business/Management", "Behavioral/Social Science", "Fine Arts"])
df2.columns=["Major", "Marks"]
df2
```

```python
fit = ols("Marks~Major", data=df2).fit()
anova = sm.stats.anova_lm(fit)
anova
```

```python
tukey = pairwise_tukeyhsd(df2['Marks'], df2['Major'])
tukey._results_table
```
