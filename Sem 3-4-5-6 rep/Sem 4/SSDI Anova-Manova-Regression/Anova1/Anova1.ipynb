```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.formula.api import ols
from statsmodels.stats.multicomp import pairwise_tukeyhsd
```

```python
df = pd.read_csv("Anova Dataset 1.csv")
df
```

```python
df2 = pd.melt(df.reset_index(), id_vars=['index'], value_vars=["Placebo", "Low Dose", "Moderate Dose"])
df2.drop(columns="index", inplace=True)
df2.columns= ["Drug", "Depression_Score"]
df2
```

```python
fit = ols("Depression_Score~Drug", data=df2).fit()
anova = sm.stats.anova_lm(fit)
anova
```

```python
tukey = pairwise_tukeyhsd(df2["Depression_Score"], groups=df2["Drug"])
tukey._results_table
```

```python
from scipy.stats import ttest_ind
ttest_ind(df['Moderate Dose'], df['Placebo'], alternative="less")
```
