```python
import os
def get_image_array(path): # converts to an array 
    image_list = os.listdir(path)
    images = []
    
    for img in image_list:
        img = cv2.imread(path + '/' + img, 0)
        img = cv2.resize(img, (255, 255))
        images.append(img)
        
    
    images = np.array(images)
    return np.array(images)
```
