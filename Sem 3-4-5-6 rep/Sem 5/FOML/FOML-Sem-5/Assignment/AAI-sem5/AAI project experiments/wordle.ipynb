```python
import random

# Sample word list (you can replace this with a more extensive knowledge base)
word_list = ["apple", "banana", "cherry", "date", "elderberry", "fig", "grape", "honeydew", "kiwi", "lemon"]

# Select a random word from the word list
secret_word = random.choice(word_list)

def is_word_valid(word):
    return len(word) == len(secret_word)

def evaluate_guess(secret, guess):
    correct_positions = sum([1 for s, g in zip(secret, guess) if s == g])
    common_letters = len(set(secret) & set(guess))
    return correct_positions, common_letters - correct_positions

def ai_wordle(word_list, max_attempts=6):
    attempts = 0
    while attempts < max_attempts:
        guess = random.choice(word_list)
        if is_word_valid(guess):
            correct, common = evaluate_guess(secret_word, guess)
            print(f"Attempt {attempts + 1}: {guess} - Correct: {correct}, Common: {common}")
            attempts += 1
            if correct == len(secret_word):
                print("Congratulations! You guessed the word:", secret_word)
                break
    if attempts == max_attempts:
        print("Out of attempts. The secret word was:", secret_word)

if __name__ == "__main__":
    ai_wordle(word_list)
```

```python
import itertools

# Define a list of valid five-letter words
word_list = ["apple", "table", "tiger", "teeth", "music", "happy", "hello", "world"]

class WordleAI():
    def __init__(self, target_word):
        self.target_word = target_word
        self.max_attempts = 6
        self.attempts = 0
        self.possible_words = word_list.copy()

    def make_guess(self):
        # In this example, the AI makes random guesses
        return self.possible_words.pop(0)

    def provide_feedback(self, guess):
        feedback = ""
        if guess == self.target_word:
            return "WordleAI wins!"
        for i in range(5):
            if guess[i] == self.target_word[i]:
                feedback += guess[i]
            elif guess[i] in self.target_word:
                feedback += guess[i].lower()
            else:
                feedback += "*"
        return feedback

    def play_game(self):
        while self.attempts < self.max_attempts:
            guess = self.make_guess()
            feedback = self.provide_feedback(guess)
            print(f"Guess {self.attempts + 1}: {guess} - Feedback: {feedback}")
            self.attempts += 1
            if guess == self.target_word:
                print("WordleAI wins!")
                break
        else:
            print("WordleAI ran out of attempts. Target word was:", self.target_word)

if __name__ == "__main__":
    target_word = "world"  # You can set the target word here
    wordle_ai = WordleAI(target_word)
    print(f"WordleAI is trying to guess: {target_word}")
    wordle_ai.play_game()
```

```python
import random

class HangmanAI:
    def __init__(self, word_list):
        self.word_list = word_list
        self.guesses = []
        self.word_to_guess = random.choice(self.word_list)
        self.remaining_attempts = 6

    def make_guess(self):
        # Implement a guessing strategy here.
        # For simplicity, we will randomly select a letter that hasn't been guessed yet.
        available_letters = [letter for letter in 'abcdefghijklmnopqrstuvwxyz' if letter not in self.guesses]
        return random.choice(available_letters)

    def update_guesses(self, guess):
        self.guesses.append(guess)

    def display_word(self):
        revealed_word = ''
        for letter in self.word_to_guess:
            if letter in self.guesses:
                revealed_word += letter
            else:
                revealed_word += '_'
        return revealed_word

    def play_game(self):
        print("Welcome to Hangman!")
        while self.remaining_attempts > 0:
            print("Word to guess:", self.display_word())
            print("Remaining attempts:", self.remaining_attempts)
            guess = self.make_guess()
            print("AI's guess:", guess)
            self.update_guesses(guess)
            if guess in self.word_to_guess:
                print("Correct guess!")
            else:
                print("Incorrect guess.")
                self.remaining_attempts -= 1
            if set(self.guesses) == set(self.word_to_guess):
                print("Congratulations, you guessed the word:", self.word_to_guess)
                break
        else:
            print("You ran out of attempts. The word was:", self.word_to_guess)

if __name__ == "__main__":
    word_list = ["apple", "banana", "cherry", "date", "fig", "grape", "kiwi", "lemon", "mango", "orange", "pear", "quince", "raspberry", "strawberry", "watermelon"]
    hangman_ai = HangmanAI(word_list)
    hangman_ai.play_game()
```
