```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_excel('PCA_case study.xlsx')
```

```python
df.interpolate(inplace=True)
```

```python
df.info()
```

```python
df = df.drop(["Model"],axis=1)
```

```python
encoded_df = pd.get_dummies(df, columns=['Manufacture'])
```

```python
encoded_df
```

```python
encoded_df.columns
```

```python
X = encoded_df[['Sales', 'Resales', 'Price', 'Engine_size', 'Horsepow',
       'WheelBase', 'Width', 'Length', 'Curb_Wgt', 'Fuel_Cap', 'Mpg',
       'Inslaes', 'Manufacture_Acura', 'Manufacture_Audi', 'Manufacture_BMW',
       'Manufacture_Buick', 'Manufacture_Cadillac', 'Manufacture_Chevrolet',
       'Manufacture_Chrysler', 'Manufacture_Dodge', 'Manufacture_Ford',
       'Manufacture_Honda', 'Manufacture_Hyundai', 'Manufacture_Infiniti',
       'Manufacture_Jeep', 'Manufacture_Lexus', 'Manufacture_Lincoln',
       'Manufacture_Mercedes-Benz', 'Manufacture_Mercury',
       'Manufacture_Mitsubishi', 'Manufacture_Nissan',
       'Manufacture_Oldsmobile', 'Manufacture_Plymouth', 'Manufacture_Pontiac',
       'Manufacture_Porsche', 'Manufacture_Saturn', 'Manufacture_Toyota',
       'Manufacture_Volkswagen']]
Y = encoded_df[['Type']]
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
from sklearn.preprocessing import StandardScaler,LabelEncoder
lda = LDA(n_components=1)
k = lda.fit(X,Y)
X_lda = lda.transform(X)
predict = lda.predict(X)
```

```python
predict
```

```python
from sklearn import metrics
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
```

```python
cm = metrics.confusion_matrix(Y,predict)
ac = metrics.accuracy_score(Y,predict)
```
