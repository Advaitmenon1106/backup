```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('death_pred.csv')
data.head()
```

```python
data.info()
```

```python
x = data.drop(columns=['DEATH_EVENT']).values
y = data['DEATH_EVENT'].values
```

```python
x.shape, y.shape
```

```python
from tensorflow import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
model = Sequential()
model.add(Dense(32, input_shape=(12, ), activation='tanh'))
model.add(Dense(16, activation='relu'))
model.add(Dense(4, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile(optimizer='adam', loss='BinaryCrossentropy', metrics=['accuracy'])
model.fit(x, y, epochs=200)
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
```

```python
lda = LDA()
model_lda = lda.fit(x, y)
```

```python
from sklearn.metrics import confusion_matrix


cf = confusion_matrix(y, model_lda.predict(x))
```

```python
accuracy = cf[0].sum()/cf.sum()
```

```python
accuracy
```

```python
from sklearn.decomposition import PCA

pca = PCA()
model_pca = pca.fit(x, y)
```

```python
model_pca.explained_variance_ratio_[0]
```

```python
model_pca.inverse_transform(x)
```

```python
model_pca.transform(x)
```

```python
model.fit(model_pca.transform(x), y, epochs=100)
```
