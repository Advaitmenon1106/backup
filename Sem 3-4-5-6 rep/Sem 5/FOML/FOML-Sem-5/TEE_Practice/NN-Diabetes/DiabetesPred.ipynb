```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('diabetes.csv')
```

```python
data.head()
```

```python
X = data.drop(columns=['Outcome']).values
y = data['Outcome'].values
```

```python
X.shape
```

```python
from tensorflow import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
model = Sequential()
model.add(Dense(32, input_shape=(X.shape[1], ), activation='tanh'))
model.add(Dense(16, activation='relu'))
model.add(Dense(4, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
model.compile(optimizer='adam', loss='BinaryCrossentropy', metrics=['accuracy'])
```

```python
model.fit(X, y, epochs=100)
```

```python
from sklearn.model_selection import train_test_split


Xtrain, Xtest, Ytrain, Ytest = train_test_split(X, y, test_size=0.3)
```

```python
model_train = Sequential()
model_train.add(Dense(32, input_shape=(Xtrain.shape[1], ), activation='tanh'))
model_train.add(Dense(16, activation='relu'))
model_train.add(Dense(4, activation='relu'))
model_train.add(Dense(1, activation='sigmoid'))
model_train.compile(optimizer='adam', loss='BinaryCrossentropy', metrics=['accuracy'])
```

```python
model_train.fit(Xtrain, Ytrain, epochs=300)
```

```python
model_train.evaluate(Xtest, Ytest)
```

```python
from sklearn.metrics import confusion_matrix


cf = confusion_matrix(Ytest, model_train.predict(Xtest).round())
```
