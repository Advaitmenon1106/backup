```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
data = load_iris()
```

```python
y = data.target
x = data.data
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA

lda = LDA()
lda_model = lda.fit(x, y)
```

```python
lda_model.predict(x)
```

```python
from sklearn.metrics import confusion_matrix


cf = confusion_matrix(y, lda_model.predict(x))
```

```python
accuracy = (cf[0][0] + cf[1][1] + cf[2][2])/cf.sum()
```

```python
accuracy
```

```python
from tensorflow import keras
from keras.layers import Dense
from keras.models import Sequential
from keras.utils import to_categorical
```

```python
yencoded = to_categorical(y)
```

```python
ann_model = Sequential(
    [
        Dense(32, input_shape=(x.shape[1], ), activation='tanh'),
        Dense(16, activation='relu'),
        Dense(4, activation='relu'),
        Dense(yencoded.shape[1], activation='sigmoid')
    ]
)
```

```python
ann_model.compile(optimizer='adam', loss='CategoricalCrossentropy', metrics=['accuracy'])
```

```python
ann_model.fit(x, yencoded, epochs=200, batch_size=35)
```
