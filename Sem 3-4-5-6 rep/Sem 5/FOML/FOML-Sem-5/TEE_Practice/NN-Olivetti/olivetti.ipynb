```python
import numpy as np
import pandas as pd
from sklearn.datasets import fetch_olivetti_faces
```

```python
data = fetch_olivetti_faces()['data']
```

```python
data.shape
```

```python
from sklearn.decomposition import PCA


pca = PCA(n_components=100)
model_pca = pca.fit(data)
```

```python
from matplotlib.pyplot import imshow
import matplotlib.pyplot as plt


imshow(model_pca.components_[0].reshape(64, 64),cmap=plt.bone())
```

```python
plt.imshow(model_pca.mean_.reshape(64, 64))
```

```python
pca2 = PCA()
model_pca_full = pca2.fit(data)
```

```python
model_pca_full.mean_.shape
```
