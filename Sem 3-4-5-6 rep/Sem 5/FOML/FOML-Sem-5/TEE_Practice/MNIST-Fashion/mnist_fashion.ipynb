```python
import numpy as np
import pandas as pd
from tensorflow import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.datasets.fashion_mnist import load_data
```

```python
(x_train, y_train), (x_test, y_test) = load_data()
```

```python
xtrain_input = x_train.reshape(x_train.shape[0], x_train.shape[1]*x_train.shape[2])
```

```python
yencoded = keras.utils.to_categorical(y_train)
Batch_Size = 1000
```

```python
model = Sequential([
    Dense(64, input_shape=(xtrain_input.shape[1], ), activation='tanh'),
    Dense(32, activation='relu'),
    Dense(16, activation='relu'),
    Dense(10, activation='softmax')
])
```

```python
model.compile(optimizer='adam', loss='CategoricalCrossentropy', metrics=['accuracy'])
```

```python
model.fit(xtrain_input, yencoded, batch_size=Batch_Size, epochs=300)
```

```python
model_eval = model.evaluate(x_test.reshape(x_test.shape[0], x_test.shape[1]*x_test.shape[2]), keras.utils.to_categorical(y_test))
```
