```python
import numpy as np
import pandas as pd
from tensorflow import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
data = pd.read_csv('BrainCancer.csv', na_values='nan').drop(columns=['Unnamed: 0']).dropna()
data.head()
```

```python
print(data['diagnosis'].unique(), data['loc'].unique(), data['ki'].unique(), data['stereo'].unique())
```

```python
len(data['time'])
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder()
categories = data[['diagnosis', 'loc', 'ki', 'stereo']].values
ohe_data = ohe.fit(categories)
```

```python
ohe_values = ohe_data.transform(categories).todense()
```

```python
ohe_values.shape
```

```python
x = np.c_[ohe_values, data['gtv'].values.reshape(-1, 1)]
```

```python
model = Sequential()
model.add(Dense(32, input_shape=(15, ), activation='tanh'))
model.add(Dense(16, activation='relu'))
model.add(Dense(4, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile(optimizer='adam', loss='BinaryCrossentropy', metrics=['accuracy'])
model.fit(x, data['status'], epochs=1000)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
model = lr.fit(np.asarray(x), data['status'])
```

```python
model.predict(np.asarray(x))
```

```python
from sklearn.metrics import confusion_matrix


confusion_matrix(data['status'], model.predict(np.asarray(x)))
```
