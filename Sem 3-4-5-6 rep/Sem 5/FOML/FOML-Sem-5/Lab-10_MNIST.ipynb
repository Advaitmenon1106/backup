```python
from sklearn.datasets import load_digits
```

```python
from sklearn import datasets  #imports the datasets module from the sklearn
digits = load_digits()
digits.data.shape
```

```python
k=digits.images[3]
k.shape
```

```python
from matplotlib import pyplot as plt
plt.imshow(k,cmap=plt.cm.bone)
```

```python
fig = plt.figure(figsize=(8, 6))
for i in range(15):
    ax = fig.add_subplot(3, 5, i + 1, xticks=[], yticks=[])
    ax.imshow(digits.images[i], cmap=plt.cm.bone)
```

```python
from sklearn.decomposition import PCA
pca = PCA()
pca.fit(digits.data)
```

```python
plt.imshow(pca.mean_.reshape(digits.images[0].shape),cmap=plt.cm.bone)
```

```python
print(pca.components_.shape)
```

```python
import numpy as np
print(pca.explained_variance_)
plt.plot(np.cumsum(pca.explained_variance_ratio_)) 
plt.show()
```

```python
plt.imshow(pca.components_[0].reshape(digits.images[0].shape),cmap=plt.cm.bone)
```

```python
##plot first 10 eigen face
fig = plt.figure(figsize=(320, 120))
for i in range(10):
    ax = fig.add_subplot(3, 100, i + 1, xticks=[], yticks=[])
    ax.imshow(pca.components_[i].reshape(digits.images[0].shape),
              cmap=plt.cm.bone)
```

```python
digitsPCA = pca.fit_transform(digits.data)
print(digitsPCA.shape)
```

```python
digits_transform = pca.inverse_transform(digitsPCA)
print(digits_transform.shape)
```

```python
plt.imshow(digits_transform[0].reshape(digits.images[0].shape),cmap=plt.cm.bone)
```
