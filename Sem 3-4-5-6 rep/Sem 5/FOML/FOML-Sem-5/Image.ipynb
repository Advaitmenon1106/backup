```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
cap = cv2.VideoCapture(0)
count = 0
while True:

 

    ret, frame = cap.read()
    count += 1
    face = cv2.resize(frame, (200,400))

 

    file_name_path = './faces/face' + str(count) + '.jpg'
    cv2.imwrite(file_name_path, face)

 

    cv2.putText(face, str(count), (50, 50), cv2.FONT_HERSHEY_COMPLEX, 1, (0,255,0), 2)
    cv2.imshow('Face Cropper', face)
        
    if cv2.waitKey(1) == 13 or count == 20: #13 is the Enter Key
        break
        
cap.release()
cv2.destroyAllWindows()
print("Collecting Samples Complete")
```

```python
img = './faces'
```

```python
import os

def get_image_array(path):
    image_list = os.listdir(path)
    images = []
    for img in image_list:
        img = cv2.imread(path + '/' + img, 0)
        img = cv2.resize(img, (400, 400))
        images.append(img)
    return np.array(images)
```

```python
img1 = get_image_array(img)
```

```python
img_reshaped = img1.reshape((img1.shape[0], img1.shape[1]*img1.shape[2]))
```
