```python
from statsmodels.tsa.stattools import adfuller
adfuller(appleStocks['Close'], autolag='AIC')
```

```python
shifted = appleStocks['Close'].shift(1)
appleStocks['Shifted_One'] = shifted-appleStocks['Close']
adfuller(appleStocks['Shifted_One'].dropna(), autolag='AIC')
```

```python
appleStocks['Shifted_One'].plot(figsize=(12, 7))
```

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
plot_pacf(appleStocks['Shifted_One'].dropna(), lags=10)
plot_acf(appleStocks['Shifted_One'].dropna(), lags=10)
```

```python
from statsmodels.tsa.arima.model import ARIMA
```

```python
model = ARIMA(appleStocks['Close'], order=(0, 1, 0)).fit()
```

```python
model.summary()
```

```python
appleStocks['Close'].iloc[1:].plot(figsize=(12, 7))
model.predict()[1:].plot(figsize=(12, 7));
```

```python
model.predict()[1:]
```
