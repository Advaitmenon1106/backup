```python
from matplotlib import pyplot
from matplotlib.pyplot import figure
from pandas import Series


np.random.seed(1)
series = np.random.normal(0, 1, 1000)
series = Series(series)

figure(figsize=(12, 7))
pyplot.plot(series)
pyplot.title('White Noise TS')
pyplot.xlabel('Time')
pyplot.ylabel('Value')
pyplot.show()
```
