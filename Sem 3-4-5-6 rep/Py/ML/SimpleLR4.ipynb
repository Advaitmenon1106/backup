```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Salary_Data.csv")
print (df)
```

```python
print (df["YearsExperience"].isna().sum())
print (df["Salary"].isna().sum())
```

```python
X = np.array(df["YearsExperience"])
Y = np.array(df["Salary"])
X = (X)/(max(X))
Y = (Y)/(max(Y))
w = 0
b = 0
LR = 0.009
it = 10000
J = []
```

```python
for x in range (it):
    Yhat = (w*X)+b
    J1 = (1/(2*len(X)))*(((Yhat-Y)**2).sum())
    J.append(J1)
    DiffWRTw = (1/float(len(X)))*((X*(Yhat-Y)).sum())
    DiffWRTb = (1/float(len(X)))*((Yhat-Y).sum())
    w = w-(LR*DiffWRTw)
    b = b-(LR*DiffWRTb)

print (w, b)
```

```python
it1 = list(range(it))
print (len(it1), len(J))
plt.plot(it1, J)
plt.show()
```
