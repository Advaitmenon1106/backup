```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("LR Dataset 2.csv")
x = df["HEIGHT"] #1= Female, 0 = male
y = df["GENDER"]
plt.scatter(x, y)
```

```python
xnorm = x/max(x)
```

```python
def sigmoid(k):
    sig = 1 / (1 + np.exp(-k))
    return sig

plt.scatter(xnorm, sigmoid(xnorm))
print(xnorm)
```

```python
LR = 0.001
it = 1000
w = 0
b = 0
sig = np.empty(len(x))
for i in range (it):
    f = (w*xnorm)+b
    DiffWRTw = (1/float(len(x)))*((f-y)*x)
    DiffWRTb = (1/float(len(x)))*((f-y))
    w = w-np.sum((LR*DiffWRTw))
    b = b-np.sum((LR*DiffWRTb))
```

```python
s = np.empty(len(x))
s = sigmoid(f)
print (f)
```

```python
plt.scatter(f, sigmoid(s))
```

```python
dfTesting = pd.read_csv("Testing LR 4.csv")
xTesting = dfTesting["HEIGHT"]
yTesting = dfTesting["GENDER"]
#xTesting = (xTesting-np.mean(xTesting))/np.std(xTesting)
xTesting = xTesting/max(xTesting)
p = (w*xTesting)+b
s = sigmoid(p)
print (s, yTesting)
```
