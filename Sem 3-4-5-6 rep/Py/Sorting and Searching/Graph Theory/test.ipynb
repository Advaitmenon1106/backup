```python
import random
```

```python
iter_dictionary = {}
```

```python
c = 0
while c<10000:
    c+=1
    numbers = []
    [numbers.append(random.randint(0, 10000)) for i in range(0, 8)]
    differences = []
    for pre_index in range(0, 8):
        for post_index in range(1, 8):
            differences.append(numbers[pre_index]-numbers[post_index])
    for number in differences:
        if abs(number)%7 == 0:
            iter_dictionary[f'iteration{(c+1)}'] = 1
            break
```
