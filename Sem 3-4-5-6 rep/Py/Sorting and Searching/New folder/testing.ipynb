```python
from GraphifyPy import UndirectedGraphs as ug
```

```python
q2Model = ug.UndirectedGraph()
```

```python
q2Model.add_node(['M115', 'M116', 'M195', 'M185', 'CS473', 'CS101', 'CS102', 'CS273'])
```

```python
ug.visualise(q2Model)
```

```python
q2Model.add_edge('M115', 'CS473')
q2Model.add_edge('M116', 'CS473')
q2Model.add_edge('M195', 'CS101')
q2Model.add_edge('M195', 'CS102')
q2Model.add_edge('M115', 'M116')
q2Model.add_edge('M115', 'M185')
q2Model.add_edge('M185', 'M195')
```

```python
ug.visualise(q2Model)
```

```python
q2ModelComplement = q2Model.complement()
q2ModelComplement.nodelist = q2Model.nodelist
```

```python
ug.visualise(q2ModelComplement)
```
