```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("ANOVA Dataset.csv")
print (df)
```

```python
df1 = df.pivot(columns="City")['Rate']
df1[1]
df1
```

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols
```

```python
lm = ols("Rate~City", data=df).fit()
anova = sm.stats.anova_lm(lm, type = 1)
print(anova)
```

```python
from statsmodels.stats.multicomp import pairwise_tukeyhsd
tukey = pairwise_tukeyhsd(df["Rate"], groups=df["City"])
tukey._results_table
```

```python
from scipy import stats
stats.ttest_ind(df)
```
