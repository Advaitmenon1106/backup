```python
import random as rand
c = 1000
y = 100
randomWalks = []
vals = []
vals.append(y)
while c>0:
    r = rand.random()
    vals.append(y)
    if r<0.5:
        randomWalks.append(-1)
        y = y-1
        vals.append(y)
    elif r>=0.5:
        randomWalks.append(1)
        y = y+1
        vals.append(y)
    c-=1
```

```python
print(randomWalks)
```

```python
import matplotlib.pyplot as plt
plt.plot(randomWalks)
```
